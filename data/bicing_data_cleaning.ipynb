## Bicing stations data cleaning
File downloaded from "https://opendata-ajuntament.barcelona.cat/data/dataset/6778b87d-da09-4eb8-9504-5c9ae5fbfad9/resource/1da43940-0750-4144-b2ff-12f420ec6026/download"

```python
import pandas as pd
import os

data = pd.read_csv("../downloads/opendatabcn_transports_bicis_punts-ancoratge-bicicletes.csv", on_bad_lines='skip',sep=';')
data.head()
```

```python
# First drop columns without values
data.drop(columns=['institution_id','institution_name','addresses_roadtype_id','addresses_roadtype_name','addresses_end_street_number','addresses_type',
    'values_id','values_attribute_id','values_category','values_attribute_name','values_value', 'values_outstanding', 'values_description'], axis=1, inplace=True)

# Drop also columns without useful info:
data.drop(columns=['register_id', 'created', 'modified', 'addresses_road_id', 'addresses_main_address', 'secondary_filters_name', 'secondary_filters_fullpath',
    'secondary_filters_tree', 'secondary_filters_asia_id', 'secondary_filters_asia_id', 'addresses_town', 'secondary_filters_id', 'geo_epgs_25831_x', 'geo_epgs_25831_y',
    'addresses_neighborhood_id','addresses_district_id'], axis=1, inplace=True)
data.head()
```

```python
# Fix address numbers 9999
def fix_street_number(num):
    try:
        if num > 2000:
                return 1
        return int(num)
    except:
        return 1
data['addresses_start_street_number'] = data['addresses_start_street_number'].apply(fix_street_number)
data.head()
```

```python
# Fix geopos
def fix_geopos(x, axis=1):
    pos = str(x).replace(".","")

    if axis==0: 
        # Some positions are wrongly spelled (4 instead of 41 for latitude)
        new_pos = f"{pos[0:2]}.{pos[2:-1]}"
    else:        
        new_pos = f"{pos[0:1]}.{pos[1:-1]}"

    return float(new_pos)

data['geo_epgs_4326_x']=data['geo_epgs_4326_x'].apply(lambda x: fix_geopos(x, axis=0))
data['geo_epgs_4326_y']=data['geo_epgs_4326_y'].apply(lambda x: fix_geopos(x, axis=1))
data.head()
```

```python
# Fix name
data["name"] = data["name"].apply(lambda row: row.split("*")[1].strip())
data.head()
```

```python
# Rename columns
data.rename(columns={'name': 'Name', 'addresses_road_name':'Street name','addresses_start_street_number':'Street number','addresses_neighborhood_name':'Neighborhood name',
    'addresses_district_name': 'District name', 'addresses_zip_code': 'Zip code', 'geo_epgs_4326_x': 'Latitud', 'geo_epgs_4326_y': 'Longitud'}, inplace=True)
data.head()
```

```python
print(data.size)
data.drop_duplicates(['Street name','Street number'], inplace=True)
data.size
```

```python
# Normalize names
import unidecode

data['Name'] = data['Name'].apply(lambda x: unidecode.unidecode(x).replace("'"," "))
data['Street name'] = data['Street name'].apply(lambda x: unidecode.unidecode(x).replace("'"," "))
data['Neighborhood name'] = data['Neighborhood name'].apply(lambda x: unidecode.unidecode(x).replace("'"," "))
data['District name'] = data['District name'].apply(lambda x: unidecode.unidecode(x).replace("'"," "))
data.head()
```

```python
#Create geojson info
def create_geojson(lon, lat):
    geojson = {
        "type":"Point",
        "coordinates":[float(lon),float(lat)]
    } 
    return geojson

data['Location'] = data.apply(lambda row: create_geojson(row["Longitud"],row["Latitud"]), axis=1)
data.drop(["Longitud", "Latitud"], axis=1, inplace=True)
data.head()
```

```python
# Upload data to MongoDB
from pymongo import MongoClient
from dotenv import load_dotenv
import os

load_dotenv()

db = MongoClient(os.getenv("URL")).core
db.bicing.insert_many(data.to_dict('records'))
```
